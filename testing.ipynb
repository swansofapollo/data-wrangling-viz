## Fetch data from the internet

```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bioinfokit import analys
import holoviews as hv 
import hvplot.pandas 
hv.extension('bokeh') 
```

```python
def download_file_to_data(url, filename):
    response = requests.get(url)
    open(filename, "wb").write(response.content)


# Download yellow taxi data
download_file_to_data(
    "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet",
    "yellow_data.parquet",
)
```

```python
yellow_df = pd.read_parquet("yellow_data.parquet")
```

```python
yellow_df
```

```python
yellow_df.dropna(inplace=True)
print(len(yellow_df))
```

```python
yellow_df.drop(['VendorID', 'RatecodeID', 'store_and_fwd_flag', 'PULocationID', 'DOLocationID','payment_type', 'extra','mta_tax', 'tolls_amount', 'improvement_surcharge','congestion_surcharge', 'airport_fee'], axis=1, inplace=True)
yellow_df.head()
```

```python
df = yellow_df.sample(n = 1000)
df.head()
```
